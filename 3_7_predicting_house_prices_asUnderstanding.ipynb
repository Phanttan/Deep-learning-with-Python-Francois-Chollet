# Boston Housing price with regression example

## Loading Data

```
from keras.datasets import boston_housing
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

## Preparing data

### Normalizing data (mean = 0, std = 1)

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_data)
train_data = scaler.transform(train_data)
train_data.mean(axis=0)
```

```
from sklearn.preprocessing import StandardScaler
train_targets = train_targets.reshape(-1,1)
scaler = StandardScaler()
train_targets = scaler.fit_transform(train_targets)
#train_targets = scaler.transform(train_targets)
train_targets.mean(axis=0)
```

## Build a network

```
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu', 
                           input_shape=(train_data.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1)) # no activation argument for linear layer because of scalar regression
    model.compile(optimizer='rmsprop', loss='mse')#, metric=['mae'])
    return model
```

## Validation using K-fold validation

```
from sklearn.model_selection import cross_val_score
from keras.wrappers.scikit_learn import KerasClassifier
neural_network = KerasClassifier(build_fn=build_model, 
                                 epochs=100,
                                 batch_size=100)
```

```
kfold_score = cross_val_score(neural_network, train_data, train_targets, cv=3, scoring='neg_mean_absolute_error',verbose=0)
```
